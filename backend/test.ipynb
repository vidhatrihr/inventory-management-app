```python
current_user = {
    'role': 'admin',
    'is_authenticated': False
}


def private():
  return 'here are my secrets'


def new_private():
  if current_user['is_authenticated'] and current_user['role'] == 'admin':
    return private()


new_private()
```

```python
from functools import wraps


def role_required(role):
  def wrapper(func):  # decorator (func => decorated_func)
    @wraps(func)
    def decorated_func(*args, **kwargs):
      if current_user['is_authenticated'] and current_user['role'] == role:
        return func(*args, **kwargs)

    return decorated_func
  return wrapper
```

```python
@role_required('admin')
def passwords(name):
  return f'here are very imp passwords.{name}'


@role_required('manager')
def private():
  return 'here are my secrets'


passwords('rahul')
private()

# smart_private = smart_maker(private, 'admin')
# smart_passwords = smart_maker(passwords, 'manager')

# smart_private()
# smart_passwords()
```
